```python
import json
import pathlib

path = sorted(pathlib.Path("../logs/spider_eval").glob("train_results_*.jsonl"))[-1]
print("Reading:", path)

records = [json.loads(line) for line in path.read_text(encoding="utf-8").splitlines()]
print(f"{len(records)} examples loaded")

for r in records:
    print("────")
    print("DB:", r["db_id"])
    print("Q :", r["question"])
    print("SQL (pred):", r["pred_sql"])
    print("Status:", r["status"])
```
